```python
loc_df = pd.read_csv('crash_data/heatmap.csv').drop(['Unnamed: 0'], axis=1)
```

```python
# create latitude and longitude squares of d x d degrees, tally up number of crashes per square
latitudes = []
longitudes = []
for i in range(len(loc_df)):
  latitudes.append(round(loc_df['LATITUDE'].iloc[i], 2)) # change decimal places to get smaller or larger sqaures
  longitudes.append(round(loc_df['LONGITUDE'].iloc[i], 2))

# crashes = []
# for i in range(len(loc_df)):
#   crash_count = len(
#       loc_df[(loc_df["LATITUDE"]==loc_df["LATITUDE"].iloc[i]) &
#           (loc_df["LONGITUDE"]==loc_df["LONGITUDE"].iloc[i])])
#   crashes.append(crash_count)

loc_df['LATITUDE'] = latitudes
loc_df['LONGITUDE'] = longitudes
# loc_df['CRASHES'] = crashes
```

```python
loc_df = loc_df.groupby(['LATITUDE', 'LONGITUDE'],as_index=False).size()
```

```python
# Remove duplicates -- only have unique longitude and latitude pairs w/ the counts.
loc_df.drop_duplicates(subset = ['LATITUDE', 'LONGITUDE'])
```

```python
loc_df = loc_df.drop(['size'], axis=1)
```

```python
loc_df["CRASHES"] = loc_df['size']
loc_df = loc_df.drop(['size'], axis=1)
```

```python
loc_df.drop(920)
```

```python
import math
log_crashes = []
for i in range(len(loc_df)):
  log_crashes.append(math.log(loc_df['CRASHES'].iloc[i]))

loc_df['LOG_CRASHES'] = log_crashes
```

```python
loc_df= loc_df.drop(['size'], axis=1)
```
