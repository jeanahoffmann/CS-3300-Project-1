```python
# Load csv files
loc_df = pd.read_csv("crash_data/2M Location Collisions.csv")
```

```python
loc_df = loc_df[['CRASH DATE', 'CRASH TIME', 'LATITUDE', 'LONGITUDE']]
loc_df = loc_df.dropna(subset=['CRASH TIME'])
loc_df['CRASH HOUR'] = pd.to_datetime(loc_df['CRASH TIME'], format='%H:%M').dt.hour
```

```python
loc_df
```

```python
loc_df = loc_df[(loc_df['LATITUDE'] != 0)]
loc_df
```

```python
loc_df = loc_df[(loc_df['LONGITUDE'] != 0)]
loc_df
```

```python
max([int(x[-4:]) for x in loc_df['CRASH DATE']])
```

```python
# filter to last 5 years
array = [(int(x[-4:]) > 2019) for x in loc_df['CRASH DATE']]
loc_df = loc_df[array]
```

```python
min([int(x[-4:]) for x in loc_df['CRASH DATE']])
```

```python
loc_df.to_csv('crash_data/barChart.csv')
```

```python
hourly_crash_count = loc_df.groupby('CRASH HOUR').size().reset_index(name='COUNT')
```
