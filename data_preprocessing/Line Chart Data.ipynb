```python
# Load csv files
loc_df = pd.read_csv("crash_data/2M Location Collisions.csv")
```

```python
loc_df = loc_df.drop((['LOCATION', 'CRASH TIME', 'BOROUGH', 'ZIP CODE','ON STREET NAME', 'CROSS STREET NAME', 'OFF STREET NAME', 'NUMBER OF PERSONS INJURED', 'NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED','NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED','NUMBER OF MOTORIST KILLED', 'CONTRIBUTING FACTOR VEHICLE 1','CONTRIBUTING FACTOR VEHICLE 2', 'CONTRIBUTING FACTOR VEHICLE 3','CONTRIBUTING FACTOR VEHICLE 4', 'CONTRIBUTING FACTOR VEHICLE 5','COLLISION_ID', 'VEHICLE TYPE CODE 1', 'VEHICLE TYPE CODE 2','VEHICLE TYPE CODE 3', 'VEHICLE TYPE CODE 4', 'VEHICLE TYPE CODE 5']), axis=1)
```

```python
loc_df
```

```python
loc_df = loc_df.dropna(axis=0)
loc_df.loc[:, 'CRASH DATE'] = pd.to_datetime(loc_df['CRASH DATE'], format='%m/%d/%Y')
loc_df
```

```python
loc_df = loc_df[(loc_df['LATITUDE'] != 0)]
loc_df
```

```python
loc_df = loc_df[(loc_df['LONGITUDE'] != 0)]
loc_df
```

```python
max([x.year for x in loc_df['CRASH DATE']])
```

```python
# filter to last 5 years
loc_df = loc_df[(loc_df['CRASH DATE'].dt.year >= 2019) & (loc_df['CRASH DATE'].dt.year <= 2024)]
```

```python
min([x.year for x in loc_df['CRASH DATE']])
```

```python
loc_df.to_csv('crash_data/lineChart.csv')
```

```python
monthly_counts = loc_df.groupby([loc_df['CRASH DATE'].dt.year.rename('Year'), loc_df['CRASH DATE'].dt.month.rename('Month')]).size().reset_index(name='Counts')
monthly_counts
```

```python
monthly_counts['Date'] = pd.to_datetime(monthly_counts['Month'].astype(str) + '-01-' + monthly_counts['Year'].astype(str), format='%m-%d-%Y')
monthly_counts
```

```python
monthly_counts.to_csv('crash_data/Crash_Monthly_Count.csv')
```
